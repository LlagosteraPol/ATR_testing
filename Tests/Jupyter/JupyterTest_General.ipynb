```python
edge_dict = {}

for edge in nx.Graph(mg).edges():
    edge_dict[edge] =  mg.number_of_edges(edge[0], edge[1])
edge_dict
```

```python
balanced = True
if max(edge_dict.values()) != min(edge_dict.values()):
    balanced = False
print('Balanced ', balanced)
```
