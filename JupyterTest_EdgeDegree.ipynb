```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.Graph([(1,4), (4,5), (4,2), (4,6), (2,3), (3,5), (6,7), (7,5), (5,8), (8,9), (9, 10), (8, 10), (10, 11), (11,12)])
```

```python
nx.draw(g, with_labels = True)
plt.show()
```

```python
edge_degree= {}
for e in g.edges:
    edge_degree[e] = g.degree(e[0]) + g.degree(e[1]) 
edge_degree
```

```python
max_degree = max(edge_degree, key=edge_degree.get)
max_degree
```

```python
min_degree = min(edge_degree, key=edge_degree.get)
min_degree
```
